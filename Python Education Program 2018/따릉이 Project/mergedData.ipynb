# 추가 컬럼 작업
## 대여요일 / 이용시간(분대) / 시간대(3시간단위) / 대여월 / 계절 / 날씨등등

```python
#  Monday == 0 … Sunday == 6 weekday
```

```python
data_all["대여요일"] = data_all["대여일시"].dt.strftime('%w')
########
시간대 = []
for i in range(int(data_all["대여일시"].size)):
    시간 = data_all['대여일시'][i].hour
    if 시간 >=0 and 시간 < 3:
        시간대.append("0~3시")
    elif 시간 >=3 and 시간 < 6:
        시간대.append("3~6시")
    elif 시간 >=6 and 시간 < 9:
        시간대.append("6~9시")
    elif 시간 >=9 and 시간 < 12:
        시간대.append("9~12시")
    elif 시간 >=12 and 시간 < 15:
        시간대.append("12~15시")
    elif 시간 >=15 and 시간 < 18:
        시간대.append("15~18시")
    elif 시간 >=18 and 시간 < 21:
        시간대.append("18~21시")
    elif 시간 >=21 and 시간 < 24:
        시간대.append("21~24시")
data_all['시간대']=시간대
data_all.head
########################
data_all["대여월"] = data_all["대여일시"].dt.strftime('%m')
```

```python
slicing =[]
def time_slice(x):
    hour_minutes = (60*int(x.strftime('%H')) + int(x.strftime('%M')))
    a = str(int(x.strftime("%Y"))%100)+"년 "+x.strftime('%m')+"월 "+x.strftime("%d")+"일 "+str(int(hour_minutes/30)+1).zfill(2) 
    slicing.append(a)
    return slicing
```

```python
for i in range(len(대여정보)):
    time_slice(대여정보["대여일시"][i])
len(slicing)
대여정보["time_slice"] = slicing
대여정보.head()
```
